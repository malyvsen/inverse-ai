```python
import pandas as pd

original = pd.read_csv('original/movie_plots.csv')
```

```python
print(f'example text: {original["Plot"][0]}')
```

```python
processed = pd.DataFrame()
processed['text'] = original['Plot']
processed['title'] = original['Title']
processed['link'] = original['Wiki Page']
```

```python
processed.head()
```

```python
def harry_potter(df):
    locations = df['title'].str.contains('Harry Potter')
    return df.loc[locations]
```

```python
harry_potter(processed)
```

```python
processed = processed.drop_duplicates(subset='link')
processed = processed.drop_duplicates(subset='title')
processed = processed.drop_duplicates(subset='text')
```

```python
harry_potter(processed)
```

```python
len(processed)
```

```python
lengths = processed['text'].map(len)
okay_length = (lengths > 256) & (lengths < 8192)
processed = processed.loc[okay_length]
```

```python
harry_potter(processed)
```

```python
len(processed)
```

```python
processed.to_csv('processed/movie_plots.csv', index=False)
```
